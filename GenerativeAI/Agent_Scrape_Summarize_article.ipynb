```python
# Scrape text from an URL

import requests
from bs4 import BeautifulSoup

def extract_article_text(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raise error if request failed

        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract paragraphs
        paragraphs = soup.find_all('p')
        article_text = '\n'.join(p.get_text() for p in paragraphs if p.get_text())

        return article_text.strip()

    except Exception as e:
        return f"Error: {e}"

# Example usage
url = 'https://www.zdnet.com/article/chatgpt-can-record-transcribe-and-analyze-your-meetings-now/'
text = extract_article_text(url)
print(text[:1000])  # Print first 1000 characters
```

```python
# Crew agent to summarize the text
# This framework can be expanded for other scenarios like creating social media post

def summarize_text(context: str) -> str:
  llm = LLM(model="gpt-4.1-mini", temperature=0.3)

  data_analyst = Agent(
      role='Educator',
      goal=f'Summarize the text provided here in bullet points. Here is the text:  {context}',
      backstory='You are a data expert',
      verbose=True,
      allow_delegation=False,
      llm=llm

  )

  test_task = Task(
      description="Summarize the given text in bullet points",

      agent=data_analyst,
      expected_output='Give a correct response'
  )

  crew = Crew(
      agents=[data_analyst],
      tasks=[test_task]
  )

  output = crew.kickoff()
```

```python
# Extract the text from the URL
url = 'https://www.zdnet.com/article/chatgpt-can-record-transcribe-and-analyze-your-meetings-now/'
context = extract_article_text(url)

#Summarize the text using the agent
summarize_text(context)
```
